```python
import tensorflow as tf
from tensorflow.keras import backend as K

def f1_score(y_true, y_pred):
    """Computes the F1 score given the true and predicted labels"""
    # Convert the labels to binary values
    y_true = K.round(y_true)
    y_pred = K.round(y_pred)

    # Compute precision and recall
    tp = K.sum(y_true * y_pred)
    fp = K.sum((1 - y_true) * y_pred)
    fn = K.sum(y_true * (1 - y_pred))
    precision = tp / (tp + fp + K.epsilon())
    recall = tp / (tp + fn + K.epsilon())

    # Compute F1 score
    f1 = 2 * precision * recall / (precision + recall + K.epsilon())
    return f1
```

```python

import numpy as np

def data_transform_pipeline(dataframe):
  dataframe = pd.DataFrame(dataframe)
  #x1-x2
  
  dataframe["Spa_Age"] = dataframe['Spa'] - dataframe['Age']
  dataframe["VRDeck_Age"] = dataframe['VRDeck'] - dataframe['Age']
  
  #x1*x2
  dataframe['FoodCourtxVRDeck'] = np.multiply(dataframe['FoodCourt'], dataframe['VRDeck'])
  #x1+x2
  dataframe['CryoSleep_+_Cabin_dech'] = np.add(dataframe["CryoSleep"], dataframe['Cabin_dech'])
  

  #sin(x1)
  for col in dataframe.iloc[:, :13].columns:
    dataframe[f'{col}_sin'] = np.sin(dataframe[col])

  #square(x1)
  for col in dataframe.iloc[:, :13].columns:
    dataframe[f'{col}_square'] = np.square(dataframe[col])

  #sqrt(x1)
  for col in dataframe.iloc[:, :13].columns:
    dataframe[f'{col}_sqrt'] = np.sqrt(dataframe[col])

  return dataframe
```
